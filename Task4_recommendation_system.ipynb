```python
import pandas as pd

movies = pd.read_csv("tmdb_5000_movies.csv")

movies.head()
```

```python
print("Dataset Shape:", movies.shape)

print("\nColumns:\n")

print(movies.columns.tolist())
```

```python
recommendation_data = movies[
    ["original_title", "overview"]
]

recommendation_data.head()
```

```python
recommendation_data.isnull().sum()
```

```python
recommendation_data = recommendation_data.dropna()

recommendation_data.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words="english"
)

movie_vectors = vectorizer.fit_transform(
    recommendation_data["overview"]
)

movie_vectors.shape

```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(
    movie_vectors
)

similarity_matrix.shape
```

```python
movie_titles = recommendation_data["original_title"]

movie_titles.head()
```

```python
def recommend_movies(movie_name):

    movie_index = movie_titles[
        movie_titles == movie_name
    ].index[0]

    similarity_scores = list(
        enumerate(
            similarity_matrix[movie_index]
        )
    )

    sorted_movies = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    print("Recommended Movies:\n")

    for movie in sorted_movies[1:6]:

        print(
            recommendation_data.iloc[
                movie[0]
            ]["original_title"]
        )
```

```python
recommend_movies("Avatar")

```

```python
recommendation_data[
    recommendation_data["original_title"].str.contains(
        "Batman",
        case=False,
        na=False
    )
][["original_title"]]
```

```python
recommend_movies("Batman Begins")
```
