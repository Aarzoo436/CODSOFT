```python
greeting_words = ["hello", "hi", "hey"]

exit_words = ["bye", "exit", "quit"]

greeting_responses = [
    "Hello",
    "Hi",
    "Hey"
]

jokes = [

    "Why don’t eggs tell jokes? Because they might crack up.",

    "Teacher: Homework kaha hai?\nStudent: Sir, network issue tha.",

    "Padhai itni karo ki padosi bole beta ab bas bhi karo.",

    "Mummy: Phone chhod do.\nMe: Bas 5 minute.\n5 hours later.",

    """Beta: Papa apki shaadi ho gayi?

Papa: Haan.

Beta: Kis se hui?

Papa: Bewkuf teri mummy se.

Beta: Wah papa ghar me hi setting kar li.""" ,

    """MAA: Beta Apple khaoge?

BETA: Nahi

MAA: Beta Mango khaoge?

BETA: Nahi

MAA: Beta Orange khaoge?

BETA: Nahi

MAA: Bilkul baap par gaya hai,
Chappal hi khayega.""" ,

    """Love aur Arrange Marriage me kya farq hai?

Love Marriage me aap apni girlfriend se shaadi karte ho.

Aur Arrange Marriage me kisi aur ki.""" ,

    """Raja aur Rani ne fix kiya ki ab baat mobile se nahi,
kabutar se karenge.

Ek din Rani ne bina khat ke kabutar uda diya.

Raja bola: Ye kya?

Rani boli: Miss call yaar.""" ,

    """Khud ko kar kanjoos itna ki har SMS bhejne se pehle
service center wale khud call karke puche,
'Sach me bhejna hai ya galti se send ho gaya tha?'"""
]

study_tips = [
    "Practice coding daily.",
    "Take short breaks while studying.",
    "Consistency is important."
]
```

```python
def greet():
    return random.choice(greeting_responses)


def current_time():
    return datetime.now().strftime("%H:%M:%S")


def study_tip():
    return random.choice(study_tips)
```

```python
def tell_joke():

    global joke_index

    if joke_index < len(jokes):

        joke = jokes[joke_index]
        

        joke_index += 1

        return joke

    return None
```

```python
name = input("Enter your name: ")

print("Welcome", name)
print("Type 'bye' to exit")
```

```python
chat_history = []

joke_index = 0
```

```python
while True:

    user_input = input(f"{name}: ").strip().lower()

    chat_history.append(user_input)

    if user_input in greeting_words:

        print("Bot:", greet())

    elif "joke" in user_input:

        joke = tell_joke()

        if joke:
            print("Bot:", joke)

        else:
            print("Bot: Mere jokes khatam ho gaye")

    elif "hasi nhi aayi" in user_input:

        joke = tell_joke()

        if joke:
            print("Bot: Accha ye wala suno")
            print(joke)

        else:
            print("Bot: Shayad mujhe comedy classes join karni padengi")

    elif "study tip" in user_input:

        print("Bot:", study_tip())

    elif "time" in user_input:

        print("Bot:", current_time())

    elif "calculate" in user_input:

        expression = user_input.replace("calculate", "")

        allowed = "0123456789+-*/(). "

        if all(ch in allowed for ch in expression):

            try:
                print("Bot:", eval(expression))

            except:
                print("Bot: Invalid calculation")

        else:
            print("Bot: Invalid calculation")

    elif "help" in user_input:

        print("""
You can ask:
- joke
- hasi nhi aayi
- time
- study tip
- calculate
""")

    elif user_input in exit_words:

        print(f"Bot: Bye {name}")
        break

    else:

        print("Bot: Samajh nahi aaya")
```

```python
print("\nCHAT HISTORY\n")

for chat in chat_history:
    print("-", chat)
```
